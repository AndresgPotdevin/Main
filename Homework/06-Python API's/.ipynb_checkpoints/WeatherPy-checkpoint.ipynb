```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from citipy import citipy

lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
lat_lngs = []
cities = []

lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    if city not in cities:
        cities.append(city)

len(cities)
```

## Perform API Calls

```python
api_key = "2df2d5a3090c8287ba418013b5494eca"

url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID="
```

```python
city_dictionary = {"city":[],
                   "lat":[],
                   "lon":[],
                   "temp":[],
                   "humidity":[],
                   "cloudiness":[], 
                   "wind_speed":[]}  
count = 0
```

```python
print('-------------------------------------------------------------')
print('Beginning Data Gathering')
print('-------------------------------------------------------------')

for city in cities:
    try:
        query_url = url + api_key + "&q=" + city
        response = requests.get(query_url)
        query = response.json()
    
        city_dictionary['city'].append(query['name'])
        city_dictionary['lat'].append(query['coord']['lat'])
        city_dictionary['lon'].append(query['coord']['lon'])
        city_dictionary['temp'].append(query['main']['temp'])
        city_dictionary['humidity'].append(query['main']['humidity'])
        city_dictionary['cloudiness'].append(query['clouds']['all'])
        city_dictionary['wind_speed'].append(query['wind']['speed'])
    
        count += 1
    
        print(f"Gathering Data for city #{str(count)} | {query['name']}")
        print(response.url)
        
    except:
        print("City not found, skipping...")
        pass

print('----------------------------------------------')
print("Data Gathering Complete")
print('----------------------------------------------')
```

```python
weather_df = pd.DataFrame.from_dict(city_dictionary)
weather_df.to_csv('city_weather_data.csv')
weather_df.head()
```

```python
plt.figure(figsize=(15,8))
plt.scatter(x=weather_df['lat'], y=weather_df['temp'], s=150, c='r', edgecolor='black')
plt.xlim(-80,100)
plt.ylim(10,120)
plt.title("Maximum Temperature Vs. Latitude")
plt.xlabel('Latitude')
plt.ylabel('Max Temperature (Farenheit)')
plt.grid(True)
plt.savefig('lat_vs_max_temp.png')
plt.show()
```
